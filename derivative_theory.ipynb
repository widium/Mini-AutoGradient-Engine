```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(-10, 10)

def power_function(X : np.array)->np.array:
	return (X**2)

Z = power_function(X)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.title("Original Value")
plt.plot(X)

plt.subplot(1, 2, 2)
plt.title("Value with Function")
plt.plot(Z)
```

```python
location = 5
Xvalue = X[location]
Zvalue = Z[location]


plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.title("Original Value")
plt.plot(X)
plt.scatter(x=location, y=Xvalue, c="green")

plt.subplot(1, 2, 2)
plt.title("Value With Function")
plt.plot(Z)
plt.scatter(x=location, y=Zvalue, c="red")
```

```python
X = np.arange(-10, 10)

def power_function(X : np.array)->np.array:
	return (X**2)

def derivative_power_function(X : np.array)->np.array:
    return (2*X)

Z = power_function(X)
dZ = derivative_power_function(X)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title("Original Value")
plt.plot(X)

plt.subplot(1, 3, 2)
plt.title("Value With Function")
plt.plot(Z)

plt.subplot(1, 3, 3)
plt.title("Derivative of Function Value")
plt.plot(dZ)
```
